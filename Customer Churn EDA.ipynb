#replacing blanks with zero as tenure is 0 and no total charges are recorded. 
#also changed data type of total charges.

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ",0)
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 0:
        return "No"
    else :
        return "Yes"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

#converted 0 and 1 values of Senior Citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x = "Churn", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

#from the given piechart we can conclude that 26.54% of our customers has been churned out. Now lets explore the reson behind it.

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3,3))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Seniorcitizens")
plt.show()
```

```python
# Percentage table
churn_pct = (
    df.groupby('SeniorCitizen')['Churn']
    .value_counts(normalize=True)
    .mul(100)
    .unstack()
)

# Plot
ax = churn_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(4,4)
)

# Add percentage labels
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f%%',
        label_type='center',
        fontsize=8
    )

plt.ylabel("Percentage (%)")
plt.title("Churn by SeniorCitizen (Percentage)")
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))
plt.xticks(rotation=0)
plt.show()
```

#comparatively a greater % of people in seniorcitizen category have been churned out

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()  # Flatten for easy indexing

for idx, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[idx], hue = df["Churn"])
    axes[idx].set_title(f'Distribution of {col}')
    axes[idx].tick_params(axis='x', rotation=45)

# Hide the last empty subplot
# axes[-1].set_visible(False)

plt.tight_layout()
plt.show()
```

#The subplots visualize count distributions across telecom service columns in a customer dataset. PhoneService shows ~70% Yes (blue bars dominant), while MultipleLines is mostly No (~65%). InternetService splits into DSL (~33%), Fiber Optic (~43%), and No (~24%); add-on services like OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport hover around 45-50% Yes with balanced No/internet-absent categories. StreamingTV and StreamingMovies both exceed 50% Yes adoption.
